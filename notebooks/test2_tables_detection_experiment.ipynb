```python
file_path = '../data/result_17102021_13101634465924/documents_files_2727_15)163ot19042021Pril5s1-17/page0.jpg'
ouput_dir_path = '../data/result_17102021_13101634465924/documents_files_2727_15)163ot19042021Pril5s1-17'
```

```python
import cv2
import numpy as np
```

```python
table_image_contour = cv2.imread(file_path, 0)
table_image = cv2.imread(file_path)
```

```python
ret, thresh_value = cv2.threshold(
    table_image_contour, 180, 255, cv2.THRESH_BINARY_INV)
```

```python
kernel = np.ones((5,5),np.uint8)
dilated_value = cv2.dilate(thresh_value,kernel,iterations = 1)
```

```python
contours, hierarchy = cv2.findContours(
    dilated_value, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    table_image = cv2.rectangle(table_image, (x, y), (x + w, y + h), (0, 0, 255), 1)
```

```python
plt.imshow(table_image)
plt.show()
cv2.namedWindow('detecttable', cv2.WINDOW_NORMAL)
```
